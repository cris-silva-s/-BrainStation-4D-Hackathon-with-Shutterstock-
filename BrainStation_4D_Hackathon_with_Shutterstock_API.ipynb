### Useful cURL parameters
- `--header` (`-H`): specifying header information with the url request. API token goes here.
- `--get` (`-G`): option to specify information to be used with the HTTP GET request.
- `--user`: specifying user information to go with the request. Format "$Consumer Key$:$Consumer Secret$"
- `--data-urlencode`: encode string in url format.

### 6. Showing some images for sanity check

One method is to get the image directly from url into memory using `scikit-image` package (a.k.a `skimage`). More information is [here](https://scikit-image.org/).

We can loop through some image urls in the `assets.preview.url` column and show them.

```python
from skimage import io

# Show some images for checking.
plt.figure(figsize=(10,10))
i = 0
while i < 9: # show only first 9 images
    # read the image directly from url.
    image = io.imread(df["assets.preview.url"][i])
    # select subplot to plot
    plt.subplot(330 + 1 + i)
    # show image
    plt.imshow(image)
    i += 1
plt.tight_layout()
plt.show()
```

## 7. Get all result pages

We saw above that the search found 7398 results. However, the actual retured result file only contain 20 images, because of the default 20 images per page (i.e. per request). We can increase the amount of results per page, for example `per_page=100` to ask for 100 images per page. The maximum allowed per page is 500. Regardless, eventually, you might need to loop through the pages to get all the results.

One way to do that is using bash `for` loop.

For example, here we will loop from page #1 to (and including) page #17. For each page I ask for 20 images.

The result of each request is saved in a separate `json` file, e.g. `dog_1.json`... `dog_17.json`

```python
%%bash

token="your token here"

for page in {1..17}
do
curl -X GET "https://api.shutterstock.com/v2/images/search" \
-H "Authorization: Bearer $token" -G \
--data-urlencode "query=pizza AND wood AND italian AND tomatoes" \
--data-urlencode "view=full" \
--data-urlencode "page=$page" \
> pizza_$page.json
done
```
